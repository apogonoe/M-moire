```python
def from_csv_to_dffreq_QLF(path):
    
    df = pd.read_csv(path, sep=',', header = 4, encoding='UTF-8')
    df.columns = ['texte','pre','lemma','post']
    df = df[['pre','post']]
    
    liste_1 = df['pre'].str.split((' ')).tolist()
    liste_2 = df['post'].str.split((' ')).tolist()
    
    liste_QLF = []
    for item in liste_1:
        for i in item:
            if i.endswith('QLF'):
                liste_QLF.append(i)
    for item in liste_2:
        for i in item:
            if i.endswith('/QLF'):
                liste_QLF.append(i)
                
    list_tuples_QLF = Counter(liste_QLF).most_common()
    
    dictionnaire_vide = dict(list_tuples_QLF)
    
    dict_items = dictionnaire_vide.items()
    
    first_hundred = list(dict_items)[:100]
    
    dico_final = dict(first_hundred)
    
    a = list(dico_final.keys())
    b = list(dico_final.values())

    df_final = pd.DataFrame(a,b)
    df_final.reset_index(inplace=True)
    df_final.columns = ["Occurrences","lemma"]  
    
    df_final['lemma'] = df_final['lemma'].str.split('/').str[0]
    
    return df_final
```

```python
def csv_tojaccard(path1,path2):
    
    df = pd.read_csv(path1, sep=",", header = 1, nrows = 100)
    df.columns = ['1','occurrence','lemme']
    df[['lemme']]
    liste1 = list(df["lemme"])
    
    df2 = pd.read_csv(path2, sep=",", header = 1, nrows = 100)
    df2.columns = ['1','occurrence','lemme']
    df2[['lemme']]
    liste2 = list(df2["lemme"])
    
    jaccard = distance.jaccard(liste1,liste2)
    
    return 1 - jaccard
```

```python
def levenshtein(seq1, seq2):
    size_x = len(seq1) + 1
    size_y = len(seq2) + 1
    matrix = np.zeros ((size_x, size_y))
    for x in range(size_x):
        matrix [x, 0] = x
    for y in range(size_y):
        matrix [0, y] = y

    for x in range(1, size_x):
        for y in range(1, size_y):
            if seq1[x-1] == seq2[y-1]:
                matrix [x,y] = min(
                    matrix[x-1, y] + 1,
                    matrix[x-1, y-1],
                    matrix[x, y-1] + 1
                )
            else:
                matrix [x,y] = min(
                    matrix[x-1,y] + 1,
                    matrix[x-1,y-1] + 1,
                    matrix[x,y-1] + 1
                )
    print (matrix)
    return (matrix[size_x - 1, size_y - 1])
```
