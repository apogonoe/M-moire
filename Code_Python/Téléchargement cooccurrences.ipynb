```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import distance
from collections import Counter

%matplotlib inline
```

```python
def from_csv_to_dffreq_VBE(path):
    
    import pandas as pd
    
    
    names = ['texte','pre','lemma','post']
    df = pd.read_csv(path, sep=',', header = 0, encoding='UTF-8', names=names)
    df = df[['pre','post']]
    
    liste_1 = df['pre'].str.split((' ')).tolist()
    for i in range(len(liste_1)):
        liste_1[i] = liste_1[i][-5:]
    liste_2 = df['post'].str.split((' ')).tolist()
    for i in range(len(liste_2)):
        liste_2[i] = liste_2[i][:5]
    
    liste_SUB = []
    for item in liste_1:
        for i in item:
            if i.endswith('/VBE'):
                liste_SUB.append(i)
    for item in liste_2:
        for i in item:
            if i.endswith('/VBE'):
                liste_SUB.append(i)
                
    list_tuples_SUB = Counter(liste_SUB).most_common()
    
    dictionnaire_vide = dict(list_tuples_SUB)
    
    dict_items = dictionnaire_vide.items()
    
    first_hundred = list(dict_items)[:100]
    
    dico_final = dict(first_hundred)
    
    a = list(dico_final.keys())
    b = list(dico_final.values())

    df_final = pd.DataFrame(a,b)
    df_final.reset_index(inplace=True)
    df_final.columns = ["Occurrences","lemma"]  
    
    df_final['lemma'] = df_final['lemma'].str.split('/').str[0]
    
    return df_final
```
