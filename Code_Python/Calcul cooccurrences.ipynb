```python
def liste_bonne_taille(liste):
    liste_vide = []
    for item in liste:
        item = item[len(item):-5]
        liste_vide.append(item)
    return liste_vide
    
```

```python
liste1_ok = liste_bonne_taille(liste_1)
liste2_ok = liste_bonne_taille(liste_2)
liste1_ok
```

```python
liste_finale = liste1_ok + liste2_ok
liste_finale
```

```python
import itertools
```

```python
liste_lemmes = list(itertools.chain.from_iterable(liste_finale))
```

```python
def from_list_todic(liste):
    dictionnaire = {}
    for word in liste:
        if word in dictionnaire:
            dictionnaire[word] = dictionnaire[word] + 1
        else:
            dictionnaire[word] = 1
            
    return dictionnaire
```

```python
dico = from_list_todic(liste_lemmes)
dico
```

```python
sortedDict = sorted(dico.items(), key=lambda x: x[1], reverse = True)
```
