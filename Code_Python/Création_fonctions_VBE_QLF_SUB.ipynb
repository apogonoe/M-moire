```python
def from_csv_to_dffreq_QLF(path):
    
    df = pd.read_csv(path, sep=',', header = 4, encoding='UTF-8')
    df.columns = ['texte','pre','lemma','post']
    df = df[['pre','post']]
    
    liste_1 = df['pre'].str.split((' ')).tolist()
    liste_2 = df['post'].str.split((' ')).tolist()
    
    liste_QLF = []
    for item in liste_1:
        for i in item:
            if i.endswith('QLF'):
                liste_QLF.append(i)
    for item in liste_2:
        for i in item:
            if i.endswith('/QLF'):
                liste_QLF.append(i)
                
    list_tuples_QLF = Counter(liste_QLF).most_common()
    
    dictionnaire_vide = dict(list_tuples_QLF)
    
    dict_items = dictionnaire_vide.items()
    
    first_hundred = list(dict_items)#[:100]
    
    dico_final = dict(first_hundred)
    
    a = list(dico_final.keys())
    b = list(dico_final.values())

    df_final = pd.DataFrame(a,b)
    df_final.reset_index(inplace=True)
    df_final.columns = ["Occurrences","lemma"]  
    
    df_final['lemma'] = df_final['lemma'].str.split('/').str[0]
    
    return df_final
```
